```python
import fitz
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
pdf_path = 'SJSTranscriptCall.pdf'
```

```python
keywords = {
    "growth": ["growth", "expansion", "increase", "revenue", "market share", "prospect", "forecast"],
    "changes": ["change", "adjustment", "shift", "transformation", "acquisition", "restructuring"],
    "triggers": ["trigger", "catalyst", "opportunity", "risk", "factor", "impact", "driver"],
    "earnings": ["earnings", "profit", "loss", "EBITDA", "cash flow", "margin", "next year", "projection"]

}
```

```python
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)
        text += page.get_text()
    doc.close()
    return text
def extract_insights(text, keywords):
    doc_nlp = nlp(text)
    insights = {category: [] for category in keywords}

    for sent in doc_nlp.sents:
        sentence_text = sent.text.lower()
        for category, category_keywords in keywords.items():
            if any(keyword in sentence_text for keyword in category_keywords):
                insights[category].append(sent.text)
    return insights
```

```python
text = extract_text_from_pdf(pdf_path)
insights = extract_insights(text, keywords)
```

```python
for category, sentences in insights.items():
    print(f"\n{category.upper()} INSIGHTS:")
    for sentence in sentences:
        print(f"- {sentence}")
```
